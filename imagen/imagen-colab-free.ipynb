```python
# Cell 2: Load the model (takes ~2 min first time)
from diffusers import StableDiffusionPipeline
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16 if device == "cuda" else torch.float32
)
pipe = pipe.to(device)
print("Model loaded!")
```

```python
# Cell 3: Generate image
prompt = "low-poly geometric meerkat illustration, muted earth tones, pastel colors"

image = pipe(
    prompt,
    num_inference_steps=30,
    guidance_scale=7.5,
).images[0]

image.save("output.png")

from IPython.display import display
display(image)
print("Saved as output.png")
```

```python
# Cell 4: Batch generate with custom prompts
prompts = [
    "low-poly geometric beaver with hammer, green hat, muted earth tones",
    "low-poly salmon fish swimming, dusty teal water background, pastel colors",
    "low-poly snow leopard sitting, mountain background, cream and sage green",
]

for i, p in enumerate(prompts):
    print(f"Generating {i+1}/{len(prompts)}: {p[:50]}...")
    img = pipe(p, num_inference_steps=30, guidance_scale=7.5).images[0]
    img.save(f"output_{i+1}.png")
    display(img)

print("All done! Files saved as output_1.png, output_2.png, output_3.png")
```

```python
# Cell 5: Download all images
from google.colab import files
import glob

for f in glob.glob("output*.png"):
    files.download(f)
print("Downloads started!")
```
